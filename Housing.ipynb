# test set 구성

```python
import numpy as np

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(housing, 0.2)
print('train_len', len(train_set), '\n', 'test_len', len(test_set))
```

데이터셋이 업데이트 될 경우를 고려해서 이전 훈련/테스트 데이터셋은 유지하는 한편 새로운 데이터를 훈련/테스트셋에 적절하게 분배하도록 하는 것이 필요하다.  
이것을 데이터의 식별자의 해시값을 기준으로 분배하는 방법이 있다.

```python
# sample의 index를 식별자로 사용한 경우

from zlib import crc32

def test_set_check(identifier, test_ratio):
    return crc32(identifier) & 0xffffffff < test_ratio * 2**32

def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(bytes(id_), test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]

housing_with_id = housing.reset_index()   # 'index' 열이 추가된 data frame
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, 'index')

print('train_len', len(train_set), '\n', 'test_len', len(test_set))
```

```python
# data의 feature를 활용해서 식별자를 만들 수도 있음
# 실제적으로 경도, 위도는 식별자로 사용하기 어려운 이유가 있는데 정밀도가 낮아 중복된 식별자가 나올 수 있기 때문이다.

housing_with_id['id'] = housing['longitude'] * 1000 + housing['latitude']
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, 'id')
```

```python
# scikit learn function을 활용하여 데이터셋을 나눌 수도 있음

from sklearn.model_selection import train_test_split

# random_state는 난수의 seed를 지정해주는 매개변수
train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)

print('train_len', len(train_set), '\n', 'test_len', len(test_set))
```

랜덤 샘플링만으로 테스트셋을 구성하면 샘플링 편향이 발생할 수 있다.<br>
특히 데이터셋의 크기가 충분히 크지 않다면 샘풀링 편향이 발생할 가능성은 더 커진다.  
테스트셋이 전체 데이터셋을 잘 대표할 수 있도록 만들기 위해서는 계층적 샘플링(stratified sampling)을 이용한다.<br>
전체 데이터셋을 계층이라는 동질의 그룹으로 나누고 각 계층의 비율에 맞춰 샘플을 추출하여 테스트셋을 구성한다.  
만약 도메인 전문가가 특정 feature가 중요하다고 했다면, 이 feature의 계층 비율을 고려하여 테스트셋을 구성해야 한다.

```python
# median income이 중요한 feature라고 가정
# 각 계층이 충분한 수의 샘플을 갖도록 계층을 나눠준다. 각 계층이 충분한 수의 샘플을 갖지 못하면 샘플링 편향이 발생할 수 있음을 유의한다.

housing['income_cat'] = pd.cut(housing['median_income'],
                              bins=[0., 1.5, 3.0, 4.5, 6.0, np.inf],
                              labels=[1, 2, 3, 4, 5])
housing['income_cat'].hist()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing['income_cat']):
    strat_train_set = housing.iloc[train_index]
    strat_test_set = housing.iloc[test_index]
    
print(strat_test_set['income_cat'].value_counts() / len(strat_test_set))
```

```python
# 샘플링 결과 비교

print('total dataset\n', housing['income_cat'].value_counts() / len(housing), '\n')
print('stratified sampling testset\n', strat_test_set['income_cat'].value_counts() / len(strat_test_set), '\n')
test_set['income_cat'] = pd.cut(test_set['median_income'],
                               bins=[0., 1.5, 3.0, 4.5, 6.0, np.inf],
                              labels=[1, 2, 3, 4, 5])
print('random sampling testset\n', test_set['income_cat'].value_counts() / len(test_set))
```

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop('income_cat', axis=1, inplace=True)
```

# 데이터 전처리

```python
# 예측변수와 레이블을 분리하여 진행

# drop()에 inplace 옵션을 안주면 드랍시킨 데이터의 복사본을 리턴함
housing = strat_train_set.drop('median_house_value', axis=1)
housing_labels = strat_train_set['median_house_value'].copy()
```

```python
# 데이터 정제: 결측치 처리

# 결측치 해결방법 3가지
# 해당 sample를 제거, 해당 feature(column)를 제거, 어떤 값으로 채움

# pandas의 dataframe method를 이용하는 경우
housing.dropna(subset=['total_bedrooms'])   # sample 제거
housing.drop('total_bedrooms', axis=1)   # feature 제거
median = housing['total_bedrooms'].median()
housing['total_bedrooms'].fillna(median, inplace=True)   # 중간값으로 채움
```

```python
# scikit learn를 이용하는 경우
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')   # imputer 추정기(estimator) instance 생성

# 중간값은 수치형 특성에만 적용가능하므로 텍스트 특성을 제외한 데이터 복사본을 생성
housing_num = housing.drop('ocean_proximity', axis=1)

# data를 이용하여 imputer 생성시 지정한 중간값(하이퍼파라메터)를 추정함(학습한다고 말함)
imputer.fit(housing_num)

print(imputer.statistics_)
print(housing_num.median().values)
```

```python
X = imputer.transform(housing_num)   # 중간값으로 결측치 채움(데이터셋을 변환). X는 numpy array
housing_tr = pd.DataFrame(X, columns=housing_num.columns, index=housing_num.index)

print(housing_tr.describe())
```

```python
# 텍스트와 범주형 특성 처리

# 범주형 특성의 경우 Integer나 one-hot vector로 변환하여 사용할 수 있다.
# 카테고리들이 순서가 있는 경우라면 정수로 변환하여 사용할 수 있지만 그렇지 않은 경우는 원핫벡터로 사용하는 것이 좋다.
# 이유는 머신러닝 알고리즘이 가까이에 존재하는 값을 비슷하다고 생각하기 때문이다.(실제로 비슷하지 않아도)
# 하지만 카테고리 수가 엄청나게 많을 경우 원핫벡터로 변환하면 입력특성이 너무 커져 학습을 느리게 할 수 있으므로
# 정수로 변환하여 사용하든지 숫자형 특성으로 재정의 하든지 해야한다.

housing_cat = housing['ocean_proximity']   # return series
housing_cat = housing[['ocean_proximity']]   # return dataframe
housing_cat.head(10)
```

```python
# Ordinal Encoder를 이용하여 integer로 변환
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
housing_cat_encoded = ordinal_encoder.fit_transform(housing_cat)   # return numpy array

print(ordinal_encoder.categories_)
print(housing_cat_encoded[:10])
```

```python
# One-Hot Encoder를 이용하여 one-hot vector로 변환
from sklearn.preprocessing import OneHotEncoder

cat_encoder = OneHotEncoder()   # sparse=False 옵션을 줄 수도 있음
housing_cat_1hot = cat_encoder.fit_transform(housing_cat)   # return scipy sparse matrix

print(housing_cat_1hot[:10])

housing_cat_1hot = housing_cat_1hot.toarray()   # 밀집행렬(dense matrix)로 변환

print(cat_encoder.categories_)
print(housing_cat_1hot)
```

마지막으로 특성 스케일링을 진행해야 한다.
대부분의 머신러닝 알고리즘은 입력 특성들의 스케일이 많이 다르면 잘 작동하지 않는다.(타깃값은 스케일링 불필요)
min-max 스케일링과 표준화(standardization)이 있다.
min-max 스케일링은 이상치에 영향을 크게 받는다.

```python
from sklearn.base import BaseEstimator, TransformerMixin

rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # *args 또는 **kargs 없음
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self  # 아무것도 하지 않습니다
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                         bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.to_numpy())
```

```python
housing_extra_attribs = pd.DataFrame(
    housing_extra_attribs,
    columns=list(housing.columns)+["rooms_per_household", "population_per_household"],
    index=housing.index)

housing_extra_attribs.head()
```

```python
# scikit learn의 pipeline은 결측치 처리, 범주형 특성 처리, 특성 스케일링을 순서대로 진행할 수 있게 함

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler())
])

housing_num_tr = num_pipeline.fit_transform(housing_num)
```

```python
# 한번에 수치형, 범주형 데이터를 변환할 수 있는 파이프라인도 있음.

from sklearn.compose import ColumnTransformer

num_attribs = list(housing_num)
cat_attribs = ['ocean_proximity']

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', OneHotEncoder(), cat_attribs)
])

# 변환시 희소행렬과 밀집행렬이 섞여있을 경우 최종행렬의 밀집정도에 따라 희소행렬이나 밀집행렬로 리턴
housing_prepared = full_pipeline.fit_transform(housing)

print(housing_prepared)
```

# 모델 학습

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(housing_prepared, housing_labels)
```

```python
some_data = housing.iloc[:5]
some_labels = housing_labels.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)
print('prediction:', lin_reg.predict(some_data_prepared))
print('label:', list(some_labels))
```

```python
from sklearn.metrics import mean_squared_error

housing_predictions = lin_reg.predict(housing_prepared)
lin_mse = mean_squared_error(housing_labels, housing_predictions)
lin_rmse = np.sqrt(lin_mse)
print(lin_rmse)
```
